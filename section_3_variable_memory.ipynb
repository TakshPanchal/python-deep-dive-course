## Garbage Collection


```python
import ctypes
import gc
```

```python
def ref_count(addr):
    return ctypes.c_long.from_address(addr).value


def get_addr(obj: int):
    return hex(id(obj))
```

```python
gc.get_objects()  # return objects
```

```python
def object_by_addr(addr):
    for obj in gc.get_objects():
        if id(obj) == addr:
            print("Object Exist in gc")
            break
    else:
        print("Object Not found in gc")
```

## Shared References and Mutability


```python
a = "object"
b = a

print(id(a))
print(id(b))
```

python uses shared reference for immutable objects for memory optimisation. Here `a` and `b` are variable holding refs to immutable objects having same value.

Both variables point to the same object addr, means that there's only one obj exist. This optimize the memory.


```python
a = 20
b = 20

print(id(a))
print(id(b))
```

with mutable objects python wont create shared references. There will be unacceptable consequences if it will be done !


```python
a = [1, 2, 3]
b = a
b.append(100)
print(a)
```

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(id(a))
print(id(b))
```

**Note** : It isnt always the case of using shared refs, dont rely on it !! sometimes python does shared refs sometime it doesnt .

Ex:


```python
a = 2000
b = 2000

print(id(a))
print(id(b))
```

#### Interning

Reusing object on demand


At startup, python(CPython) pre-loades a global list of integers in memory of range `[-5, 256]`. It is not that much large which takes time to load-up and consume lot of memory.

It is just because this range integers are common, so any time these integers are referenced, python returns the same object reference which is cached.

These integers are basically "Singleton" objects.


```python
a = -5

id(-5) == id(a)
```

```python
a = 257

id(257) == id(a)
```

```python
id(int(32)), id(int("32")), id(32), id(int("100000", 2)),
```

If you want to intern explicitly, then you must intern all the occurrence of the string

```python
def compare_without_interning(n):
    a = "a long string which is NOT interned" * 200
    b = "a long string which is NOT interned" * 200

    for _ in range(n):
        if a == b:
            pass


def compare_with_interning(n):
    a = "a long string which is interned" * 200
    b = "a long string which is interned" * 200

    for _ in range(n):
        if a is b:
            pass
```

##### Comparing whole string char by char


```python
import time

start = time.perf_counter()
compare_without_interning(10_000_000)
end = time.perf_counter()

print(end - start)
```

##### Comparing just address of the strings


```python
import time

start = time.perf_counter()
compare_with_interning(10_000_000)
end = time.perf_counter()

print(end - start)
```
